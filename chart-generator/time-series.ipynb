```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
NUM_VALUES=200
```

```python
df = pd.DataFrame({'x': list(range(NUM_VALUES)), 'y': [np.random.normal(10,10) for _ in range(NUM_VALUES)]})
df.loc[:,'moving_10'] = df['y'].rolling(window=10).mean()
df.loc[:,'moving_20'] = df['y'].rolling(window=20).mean()
df.loc[:,'moving_50'] = df['y'].rolling(window=50).mean()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(data=df,x='x',y='y',lw=2)
sns.lineplot(data=df,x='x',y='moving_10',lw=4,label='Moving n=10')
sns.lineplot(data=df,x='x',y='moving_20',lw=4,label='Moving n=20')
sns.lineplot(data=df,x='x',y='moving_50',lw=4,label='Moving n=50')
plt.savefig('moving_average.png')
```

```python
df = pd.DataFrame({'x': list(range(NUM_VALUES)), 'y': [np.random.exponential(2,1)[0] for _ in range(NUM_VALUES)]})
df
```

```python
sns.scatterplot(data=df,x='x',y='y')
sns.lineplot(x=[0,200],y=[1.8,2],color='red',lw=5)
plt.savefig('linear_regression.png')
```
